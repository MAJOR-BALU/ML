```python
def classify_features(df):
    categorical_features = []
    non_categorical_features = []
    discrete_features = []
    continuous_features = []

    for column in df.columns:
        if df[column].dtype == 'object':
            if df[column].nunique() < 10:
                categorical_features.append(column)
            else:
                non_categorical_features.append(column)
        elif df[column].dtype in ['int64']:
            if df[column].nunique() < 10:
                discrete_features.append(column)
            else:
                continuous_features.append(column)

    return categorical_features, non_categorical_features, discrete_features, continuous_features
```

```python
categorical, non_categorical, discrete, continuous = classify_features(df)
```

A pivot table is a data summarization tool used in spreadsheet programs and data analysis software, such as Microsoft Excel or pandas in Python. It allows you to summarize and aggregate large datasets into a more manageable and structured format, enabling easier analysis and interpretation.

Here are some key characteristics and functionalities of pivot tables:

1. **Aggregation**: Pivot tables aggregate and summarize data based on specified criteria. You can perform various aggregation functions such as sum, average, count, minimum, maximum, etc., on the values within the table.

2. **Dynamic Structure**: Pivot tables have a dynamic structure that allows users to easily rearrange rows and columns, change aggregation functions, and add or remove fields to analyze the data from different perspectives.

3. **Multi-level Hierarchies**: Pivot tables support multi-level row and column hierarchies, allowing you to group and organize data hierarchically for deeper analysis.

4. **Data Exploration**: Pivot tables provide an interactive way to explore and analyze complex datasets. Users can quickly drill down into details, filter data, and create visualizations to gain insights into the underlying data.

5. **Ease of Use**: Pivot tables are user-friendly and intuitive, making them accessible to users with varying levels of expertise in data analysis. They offer a graphical interface for creating and customizing tables without requiring complex coding or scripting.

6. **Customization**: Pivot tables offer extensive customization options, allowing users to format table layout, apply conditional formatting, sort data, and calculate additional derived metrics as needed.

Overall, pivot tables are powerful tools for summarizing, analyzing, and visualizing large datasets in a structured and interactive manner, making them indispensable for data analysis and decision-making in various domains such as business, finance, marketing, and research.

```python
cross_tabulation_table = pd.crosstab(df['DEATH_EVENT'], df['anaemia'])
print("Cross-tabulation of 'DEATH_EVENT' and 'anaemia':\n", cross_tabulation_table)
```

Cross-tabulation, often referred to as a contingency table or a cross-tab, is a statistical method used to analyze the relationship between two categorical variables. It presents the distribution of one variable in rows and another in columns, displaying the frequency counts or percentages of observations falling into each combination of categories.

Here's a basic explanation of cross-tabulation:

Rows: Represents the categories of one categorical variable.
Columns: Represents the categories of another categorical variable.
Cells: Displays the count or percentage of observations that fall into each combination of categories.
Cross-tabulation is particularly useful for identifying patterns and relationships between categorical variables. It helps in understanding how the distribution of one variable differs across the categories of another variable.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
X = df_new.drop('DEATH_EVENT', axis=1)
y = df_new['DEATH_EVENT']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state=42)
```

stratify=y: This parameter ensures that the splitting is done in a way that preserves the proportion of samples in each class of the target variable y. This is particularly useful for imbalanced datasets where one class may be significantly smaller than the others. Setting stratify=y helps ensure that both the training and testing sets have similar class distributions.

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

The statement `from imblearn.over_sampling import SMOTE` imports the SMOTE (Synthetic Minority Over-sampling Technique) algorithm from the `imbalanced-learn` library. Here's a breakdown:

- **`imblearn`**: This is an abbreviation for "imbalanced-learn," which is a Python library used for dealing with imbalanced datasets in machine learning. It provides various algorithms and techniques for handling class imbalance, including oversampling, undersampling, and combination sampling methods.

- **`.over_sampling`**: This part of the statement indicates that we are importing the module or submodule within the `imblearn` library that contains algorithms related to oversampling techniques.

- **`import SMOTE`**: This imports the `SMOTE` class specifically from the `over_sampling` module. SMOTE is a popular oversampling technique used to balance the class distribution by generating synthetic samples of the minority class.

In summary, `from imblearn.over_sampling import SMOTE` imports the SMOTE algorithm from the `imbalanced-learn` library, allowing us to use it to address class imbalance by oversampling the minority class in our machine learning dataset.

```python
log_reg_model = LogisticRegression()
log_reg_model.fit(X_train_resampled, y_train_resampled)
```

```python
log_reg_pred = log_reg_model.predict(X_test)
log_reg_accuracy = accuracy_score(y_test, log_reg_pred)
print("Logistic Regression Accuracy:", log_reg_accuracy)
```

```python
y_pred_train = log_reg_model.predict(X_train)
train_accuracy = accuracy_score(y_train, y_pred_train)
print("Training Accuracy:", train_accuracy)
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(X_train_resampled, y_train_resampled)
```

```python
rf_pred = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)
print("Random Forest Accuracy:", rf_accuracy)
```

```python
y_pred_train = rf_model.predict(X_train)
train_accuracy = accuracy_score(y_train, y_pred_train)
print("Training Accuracy:", train_accuracy)
```

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train_resampled, y_train_resampled)
```

```python
dt_pred = dt_model.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_pred)
print("Decision Tree Accuracy:", dt_accuracy)
```

```python
y_pred_train = dt_model.predict(X_train)
train_accuracy = accuracy_score(y_train, y_pred_train)
print("Training Accuracy:", train_accuracy)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train_resampled[['creatinine_phosphokinase', 'ejection_fraction', 'serum_sodium', 'time']])
X_test_scaled = scaler.transform(X_test[['creatinine_phosphokinase', 'ejection_fraction', 'serum_sodium', 'time']])
```

StandardScaler is a preprocessing technique used to standardize features by removing the mean and scaling to unit variance. It is a part of the scikit-learn library in Python.

Here's what StandardScaler does:

It computes the mean and standard deviation of each feature in the training data.
It then subtracts the mean from each feature and divides by the standard deviation, effectively centering the data around 0 and scaling it to have unit variance.
StandardScaler preserves the shape of the original distribution of the data while ensuring that each feature has the same scale.
StandardScaler is particularly useful when:

Features have different units of measurement.
Features have different ranges.
Features follow a normal distribution or approximately so.
Using StandardScaler can improve the performance of certain machine learning algorithms, such as support vector machines, logistic regression, and k-nearest neighbors, which rely on the assumption of having standardized features.

```python
y_pred_train = svm_model.predict(X_train_scaled)
train_accuracy = accuracy_score(y_train_resampled, y_pred_train)
print("Training Accuracy:", train_accuracy)
```
